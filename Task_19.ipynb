```python
def histplot():
  plt.subplot(2,2,1)
  sns.histplot(df['Price'])
  plt.title('Price')

  plt.subplot(2,2,2)
  sns.histplot(df['number of bathrooms'])
  plt.title('number of bathrooms')

  plt.subplot(2,2,3)
  sns.histplot(df['condition of the house'])
  plt.title('Condition of nthe House')

  plt.subplot(2,2,4)
  sns.histplot(df['grade of the house'])
  plt.title('Grade of the house')
  plt.tight_layout()

histplot()
```

```python
# Insights
# Price distribution is positively skewed.
# Area has wide variation.
# Bedrooms are concentrated around common values.
# Bathrooms mostly range between 1–4.
```

```python
def countplot():
  plt.subplot(2,2,1)
  sns.countplot(data =df,x='waterfront present')
  plt.title('Waterfront present ')

  plt.subplot(2,2,2)
  sns.countplot(data =df,x='number of floors')
  plt.title('Number of flors')
  plt.tight_layout()
countplot()
```

```python
# Insights
# Some categories have more houses than others.
# Certain locations dominate the dataset.
# Highly populated categories have more training samples.
# Rare categories may have less impact during modeling.
```

```python
numeric_cols =["Price","number of bedrooms","number of bathrooms","number of floors","Built Year"]
corr = df[numeric_cols].corr(numeric_only =True)
sns.heatmap(corr,annot =True)
print("Correlation Heatmap of Numeric Features")
```

```python
# Observations
# Area is positively correlated with Price.
# Bedrooms and bathrooms show moderate positive correlation.
# Some variables have weak correlation.
# Highly correlated features are useful for prediction.
```
