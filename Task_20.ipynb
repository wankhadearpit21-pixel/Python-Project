```python
import pandas as pd
```

```python
df = pd.read_csv("/content/ford_car_dataset.csv")
```

```python
X =df.drop("price",axis =1)
Y =df["price"]

print("Shape of X:",X.shape)
print("Shape of Y :",Y.shape)
```

```python
cat_cols = X.select_dtypes(include="object").columns
print("Categorical Columns :")
print(cat_cols)
```

```python
X =pd.get_dummies(X,columns =cat_cols).astype(int)
print(X.head())
```

```python
from sklearn.preprocessing import StandardScaler

num_cols = ["year","mileage","tax","mpg","engineSize"]

scaler = StandardScaler()
X[num_cols] = scaler.fit_transform(X[num_cols])

print(X.head())
```

```python
from  sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test =train_test_split(X,Y,test_size =0.33,random_state =42)

print("X_train  Shape :",X_train.shape)
print("X_test  Shape :",X_test.shape)
print("y_train  Shape :",y_train.shape)
print("y_test  Shape :",y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

# Create Model
model = LinearRegression()

# Train Model
model.fit(X_train,y_train)
```

```python
y_pred =model.predict(X_test)

comparision = pd.DataFrame({"Actual":y_test.values,"Predicated":y_pred})

print(comparision.head(10))
```

```python
from sklearn.metrics import r2_score

r2 =r2_score(y_test,y_pred)

print("R2 Score :",r2)
```

```python
import joblib
joblib.dump(model,"LR_ford_car.pkl")

joblib.dump(scaler,"scaler.pkl")

joblib.dump(X.columns.tolist(),"columns.pkl")

print("Files Are Saved Successfully")
```
