```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./House Price India.csv")
df
```

```python
x = df.drop('Price',axis=1)
y = df['Price']
print(" Independent Features (X) :")
print(x.columns)
print("\nDependent Feature (Y) :")
print(y.name)
print("\nShape of x :")
print(x.shape)
print("Shape of y :")
print(y.shape)
```

```python
cat_cols = df.select_dtypes('object').columns
print("Categorical columns :")
print(cat_cols)

num_cols = df.select_dtypes('int','float').columns
print("\nNumerical columns :")
print(num_cols)
num_cols = num_cols.drop('Price')
```

```python
encoded = pd.get_dummies(df)
encoded.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[num_cols])
scaled_df = pd.DataFrame(scaled_features, columns=num_cols)
scaled_df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.model_selection import train_test_split

y = df['Price']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print(y_pred[:10])

print(y_test.iloc[:10])
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print("R² Score:", r2)
```

```python
import joblib

# Save the trained Linear Regression model
joblib.dump(model, "House Price India.pkl")

# Save the StandardScaler object
joblib.dump(scaler, "scaler.pkl")

# Save the list of feature columns
joblib.dump(num_cols.tolist(), "columns.pkl")

print("Model and preprocessing objects saved successfully!")
```
